```python
iris_filename = '/content/Iris.csv'
iris = pd.read_csv(iris_filename)
```

```python
iris.info()
```

```python
#Melakukan operasi aritmatika
#Hitung perbandingan antara panjang dan lebar sepal untuk setiap target, dan
iris['sepal_ratio'] = iris['SepalLengthCm']/iris['SepalWidthCm']
print (iris.head())
```

```python
# Mengginakan .apply() untuk operasi kostum Hitung luas petal (panjang petal x lebar petal) untuk setiap target
# dan tambahkan DataFrame sebagai kolom baru
# axis = 0 itu baris. axis = 1 itu kolom
iris['petal_area'] = iris.apply(lambda row: row['PetalLengthCm'] * row['PetalWidthCm'], axis = 1)
print (iris.head())
```
